```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv("canada_income.csv")
df
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("canada income")
plt.scatter(df.year,df.canada_income,color="red", marker ="+")
```

```python
new_df =df.drop("canada_income", axis ="columns")
new_df
```

```python
income = df.canada_income
income
```

```python
model = LinearRegression()
model.fit(new_df,income)
```

```python
model.predict([[2020]])
```

```python
model.coef_
```
